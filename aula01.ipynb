```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return x**2 - 1
```

```python
x = np.linspace(-3, 3, 1000)
y = function(x)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y, label='f(x) = x^(-1)', color='blue')
plt.axhline(-1, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
# plt.grid()

plt.show()
```
