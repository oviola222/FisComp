```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 16})
```

```python
# Define f(x) and take ∂ₓf(x) using the finite difference method

def funcao(x, h=1e-10):
    f = x**2 - 1   
    df = ( ((x + h)**2 - 1) - ((x)**2 - 1) ) / h  

    # f = (1/2)*x**4 - 4*x**2
    # df = ( (1/2)*(x + h)**4 - 4*(x + h)**2 - ((1/2)*x**4 - 4*x**2) ) / h
    return f, df
```

```python
x = np.linspace(-3, 3, 1000)             # x linspace
y = np.array([funcao(i)[0] for i in x])  # f(x)
```

```python
# Define the gradient descent function parameters

alpha = 0.8     # learning rate
x_0 = 3          # initial point 
steps = [x_0]    # create a list to store the steps
n = 100           # number of steps
```

```python
# Gradient descent algorithm

for _ in range(n):
    _ , grad = funcao(x_0)  # get ∇f(x) = ∂ₓf
    x_0 +=  - alpha * grad
    steps.append(x_0)
```

```python
plt.figure(figsize=(6, 6))

plt.plot(x, y, lw=3, label=r'$f(x)$')

plt.plot(steps, np.array([funcao(i)[0] for i in steps]), c='r', ls=':', lw=2)
plt.scatter(steps, np.array([funcao(i)[0] for i in steps]), c='r', marker='o', s=50, label=r'$\Delta\vec{x}$', zorder = 99)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(lw=0.5)

# get the minimum point
x_min = steps[-1]         # last point in the steps
y_min = funcao(x_min)[0]  # get f(x_min)
print(f"Estmated minimum point: (x_min, y_min) = ({x_min:.1f}, {y_min:.1f})")

plt.tight_layout()
plt.show()
```
