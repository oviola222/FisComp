```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

plt.rcParams['font.size'] = 16
```

# Example 1: $f(x)=\sin x$

```python
# 1. Generate Training Data
np.random.seed(42)  # for reproducibility
num_samples = 100
angles_train = np.random.uniform(0, 4 * np.pi, num_samples).reshape(-1, 1)  # Reshape to (n_samples, n_features)
sin_values_train = np.sin(angles_train)

# Add some noise to the training data (optional, but can make it more realistic)
noise = np.random.normal(0, 0.1, sin_values_train.shape)
sin_values_train += noise

# 2. Define and Train the FCNN Model
# MLPRegressor expects input features to be 2D arrays (n_samples, n_features)
# For a simple angle input, n_features is 1

# Define the model
model = MLPRegressor(
    hidden_layer_sizes=(10,10,10),  # One hidden layer with 10 neurons
    activation='tanh',         # Rectified Linear Unit activation function
    solver='adam',             # Optimization algorithm
    max_iter=100000,             # Maximum number of iterations
    random_state=42,            # For reproducibility,
    learning_rate_init = 0.001,
    tol = 1e-8
)

# Train the model
model.fit(angles_train, sin_values_train)

# 3. Generate Test Data for Evaluation
num_test_samples = 50
angles_test = np.linspace(0, 6 * np.pi, num_test_samples).reshape(-1, 1)
sin_values_true = np.sin(angles_test)

# 4. Make Predictions
sin_values_predicted = model.predict(angles_test)

# 5. Evaluate the Model (Optional, but good practice)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(sin_values_true, sin_values_predicted)
print(f"Mean Squared Error on Test Data: {mse}")
```

```python
# 6. Visualize the Results
plt.figure(figsize=(6, 5))
plt.scatter(angles_train, sin_values_train, label='Training Data', alpha=0.5)
plt.plot(angles_test, sin_values_true, label=r'True sin($\theta$)', color='blue')
plt.plot(angles_test, sin_values_predicted, label=r'Predicted sin($\theta$)', color='red')
plt.xlabel(r'$\theta$ (rad)')
plt.ylabel(r'sin($\theta$)')
plt.title(r'FCNN Interpolation of sin($\theta$)')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

# Example 2: $f(x)=\sin x/x$

```python
# 1. Generate Training Data
np.random.seed(42)  # for reproducibility
num_samples = 100
angles_train = np.random.uniform(0, 2 * np.pi, num_samples).reshape(-1, 1)  # Reshape to (n_samples, n_features)
sync_values_train = np.sinc(angles_train)

# Add some noise to the training data (optional, but can make it more realistic)
noise = np.random.normal(0, 0.05, sync_values_train.shape)
sync_values_train += noise

# 2. Define and Train the FCNN Model
# MLPRegressor expects input features to be 2D arrays (n_samples, n_features)
# For a simple angle input, n_features is 1

# Define the model
model = MLPRegressor(
    hidden_layer_sizes=(5, 5, 5),  # One hidden layer with 10 neurons
    activation='tanh',              # Rectified Linear Unit activation function
    solver='adam',                  # Optimization algorithm
    max_iter=100000,                # Maximum number of iterations
    random_state=42,                # For reproducibility,
    learning_rate_init = 0.0001,
    tol = 1e-10
)

# Train the model
model.fit(angles_train, sync_values_train)

# 3. Generate Test Data for Evaluation
num_test_samples = 50
angles_test = np.linspace(0, 4 * np.pi, num_test_samples).reshape(-1, 1)
sync_values_true = np.sinc(angles_test)

# 4. Make Predictions
sync_values_predicted = model.predict(angles_test)

# 5. Evaluate the Model (Optional, but good practice)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(sync_values_true, sync_values_predicted)
print(f"Mean Squared Error on Test Data: {mse}")
```

```python
# 6. Visualize the Results
plt.figure(figsize=(6, 5))
plt.scatter(angles_train, sync_values_train, label='Training Data', alpha=0.5)
plt.plot(angles_test, sync_values_true, label=r'True sync($\theta$)', color='blue')
plt.plot(angles_test, sync_values_predicted, label=r'Predicted sync($\theta$)', color='red')
plt.xlabel(r'$\theta$ (rad)')
plt.ylabel(r'sync($\theta$)')
plt.title(r'FCNN Interpolation of sync($\theta$)')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

# Example 3: $f(x)=e^{-x^2}$

```python
# 1. Generate Training Data
np.random.seed(42)  # for reproducibility
num_samples = 100
data_train = np.random.uniform(-10, 10, num_samples).reshape(-1, 1)  # Reshape to (n_samples, n_features)
gauss_values_train = np.exp(-(data_train)**2)

# Add some noise to the training data (optional, but can make it more realistic)
noise = np.random.normal(0, 0.01, gauss_values_train.shape)
gauss_values_train += noise

# 2. Define and Train the FCNN Model
# MLPRegressor expects input features to be 2D arrays (n_samples, n_features)
# For a simple angle input, n_features is 1

# Define the model
model = MLPRegressor(
    hidden_layer_sizes=(10, 10),  # One hidden layer with 10 neurons
    activation='tanh',              # Rectified Linear Unit activation function
    solver='adam',                  # Optimization algorithm
    max_iter=100000,                # Maximum number of iterations
    random_state=42,                # For reproducibility,
    learning_rate_init = 0.001,
    tol = 1e-10
)

# Train the model
model.fit(data_train, gauss_values_train)

# 3. Generate Test Data for Evaluation
num_test_samples = 50
data_test = np.linspace(-15, 15, num_test_samples).reshape(-1, 1)
gauss_values_true = np.exp(-(data_test)**2)

# 4. Make Predictions
gauss_values_predicted = model.predict(data_test)

# 5. Evaluate the Model (Optional, but good practice)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(gauss_values_true, gauss_values_predicted)
print(f"Mean Squared Error on Test Data: {mse}")
```

```python
# 6. Visualize the Results
plt.figure(figsize=(6, 5))
plt.scatter(data_train, gauss_values_train, label='Training Data', alpha=0.5)
plt.plot(data_test, gauss_values_true, label=r'True $e^{-x^2}$', color='blue')
plt.plot(data_test, gauss_values_predicted, label=r'Predicted $e^{-x^2}$', color='red')
plt.xlabel(r'$x$ (1)')
plt.ylabel(r'$e^{-x^2}$')
plt.title(r'FCNN Interpolation of $e^{-x^2}$')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
